```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import seaborn as sns
```

```python
min_weight, max_weight, n_weights = 0, 1, 21
weights_of_text_features = np.linspace(min_weight, max_weight, n_weights)
```

```python
# Get result
mAPs = np.load("./outputs/xp_weights.npy")
```

```python
sns.set_theme()
f = plt.figure(figsize=(15,6), dpi=300)
plt.plot(weights_of_text_features, mAPs)
plt.xlabel('Weights of text features (%)')
plt.ylabel('mAP')
for ax in f.axes:
  ax.xaxis.set_major_formatter(PercentFormatter(1))
plt.show()
f.savefig('./outputs/xp_weights.jpg')
```

```python
zoomed_slice = slice(1, 10)

f = plt.figure(figsize=(15,6), dpi=300)
plt.plot(weights_of_text_features[zoomed_slice], mAPs[zoomed_slice])
plt.xlabel('Weights of text features (%)')
plt.ylabel('mAP')
for ax in f.axes:
  ax.xaxis.set_major_formatter(PercentFormatter(1))
plt.show()
f.savefig('./outputs/xp_weights_zoomed.jpg')
```

```python
pd.DataFrame([
  (weight, mAP)
  for weight, mAP in zip(weights_of_text_features, mAPs)
], columns=['Weight (%)', 'mAP']).to_excel('./outputs/xp_weights.xlsx', engine='openpyxl')
```
