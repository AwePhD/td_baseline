```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get grid
min_threshold, max_threshold, n_threshold = .1, .8, 4
min_weight, max_weight, n_weight = 0, 1, 11
threshold_grid, weight_grid = np.meshgrid(
    np.linspace(min_threshold, max_threshold, n_threshold),
    np.linspace(min_weight, max_weight, n_weight),
    sparse=True
)
```

```python
# Get result
mAPs = np.load("outputs/xp_threshold_weight.npy")
```

```python
%matplotlib widget
ax = plt.figure().add_subplot(projection="3d")
ax.invert_yaxis()
ax.plot_surface(
    threshold_grid, weight_grid, mAPs,
)
ax.set(
    xlabel = 'Eval threshold',
    ylabel = 'Weight Text / Image',
    zlabel = 'mAP',
)
ax.view_init(azim=-45)
plt.show()
```
